```python
# Set upper and lower bounds based on percentiles
lower_bound = APS['MRP'].quantile(0.01)
upper_bound = APS['MRP'].quantile(0.99)
```

```python
APS['MRP'] = APS['MRP'].clip(lower=lower_bound, upper=upper_bound)

print(APS)
```

```python
# Replace outliers with the median of the column
median_value = APS['MRP'].median()
APS['MRP'] = APS['MRP'].apply(lambda x: median_value if x > upper_bound or x < lower_bound else x)

print(APS)
```

```python
# Apply log transformation
APS['MRP'] = np.log1p(APS['MRP'])

print(APS)
```
